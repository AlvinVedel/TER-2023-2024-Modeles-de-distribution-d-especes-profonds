```python
import pandas as pd
```

```python
blind = pd.read_csv("../Data/test_blind.csv", sep=";") #fichier csv qui sera merge pas les prédiction et soumis sur Kaggle
presences = pd.read_excel("../Data/Presences_Absences_train.xlsx") #fichier csv presences/absences
```

```python
blind
```

```python
nb_especes = presences["speciesId"]
nb_especes = nb_especes.to_frame()
nb_especes['count'] = 1
sumup = nb_especes.groupby('speciesId').sum()
sumup = sumup.sort_values("count", ascending=False)
# On compte le nombre d'espèce et leur occurence 
sumup
```

```python
top_50= []
# On extrait les 50 plus présentes et on récupère leurs Ids
i=0
while i< 50 :
    value = sumup.iloc[i]["count"]
    top_50 += sumup[sumup["count"]==value].index.tolist()
    i+=1
```

```python
predict = ''
for i in top_50 :
    predict +=str(i)+ " "
predict
```

```python
most_present = blind.copy()
most_present["Predicted"] = predict
most_present
```

```python
most_present_test = most_present[["Id", "Predicted"]]
most_present_test
```

```python
most = most_present_test.groupby(by="Id").sum()
most
```

```python
most.to_csv("most_present_test2.csv")
```

```python
import random
def random_species(df):
    random_5= []
            # On extrait les 5 valeurs au hasard et on récupère leurs Ids
    i=0
    nb_species = len(df.index)
    while i< 5 :
        alea = random.randint(0, nb_species-1) 
        value = df.iloc[alea]
        if value.name not in random_5 :  
            random_5.append(value.name)
            i+=1

    predict = ''
    for i in random_5 :
        predict +=str(i)+ " "
    return predict
# fonction qui retourn 5 espèce aléatoires
```

```python
random_species(sumup)
```

```python
random_test = blind.copy()
random_sp = []
for i in range (0, len(random_test.index)) :
    random_sp.append(random_species(sumup))

    
random_test["Predicted"] = random_sp
random_test
```

```python
random = random_test[['Id', 'Predicted']]
random
```

```python
randoms = random.groupby(by="Id").sum()
randoms.to_csv("random_test.csv")
```
