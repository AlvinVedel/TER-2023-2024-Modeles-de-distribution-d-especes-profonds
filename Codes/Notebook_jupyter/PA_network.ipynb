```python
import pandas as pd
import numpy as np
import keras
import tensorflow as tf
from keras import layers
from tensorflow.keras.layers import Dense
import random
from keras.initializers import glorot_uniform
print("imports OK")
```

```python
df = pd.read_csv("../../our_data/data_ter_distribution/outputed_csv/all_infos.csv", index_col=0)
df_test = pd.read_csv("../../our_data/data_ter_distribution/outputed_csv/test_all_infos.csv", index_col=0)
print("dfs lu")
```

```python
X = df.iloc[:,0:51]
y = df.iloc[:, 51:]
X = X.drop(columns='surveyId')
X = X.drop(columns="geoUncertaintyInM")
X = X.drop(columns="HumanFootprint-Roads")


print("X et y créées")
```

```python
matrix = []
for i in range(len(X)):
    new_liste = []
    for col in X.columns :
        new_liste.append(X[col][i])
    matrix.append(new_liste)

yt = y.copy()
yt = np.array(yt).astype(np.float32)
Xt = np.array(matrix)
Xt = Xt.astype(np.float32)

print("conversion en array")
```

```python
model = keras.Sequential()
model.add(keras.Input(shape=(48,)))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))

model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))

model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))

model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))
model.add(layers.Dense(500, activation="relu"))


model.add(layers.Dense(5016, activation="sigmoid", name='output'))

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss="binary_crossentropy"
  )

print("création + compilation modèle")
print(model.summary())
```

```python
history = model.fit(
    Xt,
    yt,
    batch_size=10,
    epochs=5,
)
print("Entrainement terminé")
```
