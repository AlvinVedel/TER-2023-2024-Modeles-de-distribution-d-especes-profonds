```python
import pandas as pd
import numpy as np
#import keras
#import tensorflow as tf
#from keras import layers
#from tensorflow.keras.layers import Dense
#import random
print("imports OK")
```

```python
df = pd.read_csv("../../our_data/output/PresenceAbsenceSurveys/GLC24-PA-metadata-train.csv")
bios = pd.read_csv("../../our_data/output/EnvironmentalRasters/Climate/GLC24-PA-train-bioclimatic-average.csv")
bios = bios.fillna(bios.mean())
elev = pd.read_csv("../../our_data/output/EnvironmentalRasters/Elevation/GLC24-PA-train-elevation.csv")
elev = elev.fillna(elev.mean())
human = pd.read_csv("../../our_data/output/EnvironmentalRasters/HumanFootprint/GLC24-PA-train-human-footprint.csv")
human = human.fillna(human.mean())
land = pd.read_csv("../../our_data/output/EnvironmentalRasters/LandCover/GLC24-PA-train-landcover.csv")
land = land.fillna(land.mean())
soil = pd.read_csv("../../our_data/output/EnvironmentalRasters/Soilgrids/GLC24-PA-train-soilgrids.csv")
soil = soil.fillna(soil.mean())

print("Lecture dfs OK")

df_final = pd.merge(df, bios, on="surveyId", how="inner")
df_final = pd.merge(df_final, elev, on='surveyId', how='inner')
df_final = pd.merge(df_final, human, on='surveyId', how='inner')
df_final = pd.merge(df_final, land, on='surveyId', how='inner')
df_final = pd.merge(df_final, soil, on='surveyId', how='inner')

print("Fusion OK")
```

```python
aggregations = {'speciesId': lambda x: list(x)}  # Colonnes avec des listes
for col in df_final.columns:
    if col != 'surveyId' and col != 'speciesId':
        aggregations[col] = 'first'
        
unique_df = df_final.groupby('surveyId').agg(aggregations).reset_index()

print("Aggregation OK")
```

```python
liste_espece = list(set(df_final["speciesId"].to_list()))

columns = []
for j in range(len(liste_espece)):
    print(j)
    column = []
    for i in range(len(unique_df)):
        if liste_espece[j] in unique_df["speciesId"][i] :
            column.append(1)
        else:
            column.append(0)
    columns.append(column)

all_infos_df = unique_df.copy()
for i in range(len(columns)):
    all_infos_df[int(liste_espece[i])] = columns[i]
    
print("matrice O 1 OK")
```

```python
all_infos_df = all_infos_df.drop(columns=["speciesId", "region", 'country', 'areaInM2'])
print("colonnes supprimées")
```

```python
all_infos_df.to_csv("../../our_data/data_ter_distribution/outputed_csv/all_infos.csv", sep=',')

print("csv outputed")
```
