```python
import random
def random_species(liste, nb):
    random_nb= []  # On extrait les n valeurs au hasard et on récupère leurs Ids
    copied_list = liste.copy()
    while len(random_nb) < nb :
        nb_species = len(copied_list)
        alea = random.randint(0, nb_species-1) 
        value = copied_list[alea]
        del copied_list[alea]
        random_nb.append(value)
    return random_nb
# fonction qui retourn nb espèce aléatoires

def as_chaine(liste) :
    i=0
    chaine = ""
    while i<len(liste)-1 :
        chaine=chaine+str(liste[i])+' '
        i+=1
    chaine+=str(liste[len(liste)-1])
    return chaine
```

```python
def csv_k_plus_present(k, presences, blind):
    # On compte le nombre d'espèce et leur occurence 
    nb_especes = presences["speciesId"]
    nb_especes = nb_especes.to_frame()
    nb_especes['count'] = 1
    sumup = nb_especes.groupby('speciesId').sum()
    sumup = sumup.sort_values("count", ascending=False)
    
    top_k= []
    # On extrait les k plus présentes et on récupère leurs Ids
    i=0
    while i< k :
        value = sumup.iloc[i]["count"]
        top_k += sumup[sumup["count"]==value].index.tolist()
        i+=1
    
    predict = ''
    for i in top_k :
        predict +=str(i)+ " "
    
    most_present = blind.copy()
    most_present["Predicted"] = predict
    most_present = most_present[["Id", "Predicted"]]
    most_present = most_present.groupby(by="Id").sum()
    output_filename = f"{k}_plus_present.csv"
    most_present.to_csv(output_filename)
```

```python
csv_k_plus_present(30, presences, blind)
```
