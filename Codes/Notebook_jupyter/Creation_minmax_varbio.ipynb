```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import NearestNeighbors
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../../Data/Data_train/abiodata_speciesID.csv", sep =";")
```

```python
resultats_min = data.groupby('speciesId')[['bio1', 'bio2', 'bio3','bio4', 'bio5', 'bio6','bio7', 'bio8',
                                          'bio9', 'bio10', 'bio11','bio12', 'bio13', 'bio14','bio15', 'bio16',
                                          'bio17', 'bio18', 'bio19']].min()
resultats_max = data.groupby('speciesId')[['bio1', 'bio2', 'bio3','bio4', 'bio5', 'bio6','bio7', 'bio8',
                                          'bio9', 'bio10', 'bio11','bio12', 'bio13', 'bio14','bio15', 'bio16',
                                          'bio17', 'bio18', 'bio19']].max()
```

```python
# Renommer les colonnes 1 à 19 en ajoutant le préfixe 'max'
nouvelles_colonnes = {}
for i in range(1, 20):
    ancien_nom = f'bio{i}'  # Remplacez 'col' par le nom réel de vos colonnes si nécessaire
    nouveau_nom = f'max{ancien_nom}'
    nouvelles_colonnes[ancien_nom] = nouveau_nom

resultats_max = resultats_max.rename(columns=nouvelles_colonnes)

nouvelles_colonnes = {}
for i in range(1, 20):
    ancien_nom = f'bio{i}'  # Remplacez 'col' par le nom réel de vos colonnes si nécessaire
    nouveau_nom = f'min{ancien_nom}'
    nouvelles_colonnes[ancien_nom] = nouveau_nom

resultats_min = resultats_min.rename(columns=nouvelles_colonnes)
resultats = pd.concat([resultats_min, resultats_max], axis=1)
resultats
```

```python
resultats.to-csv('../../Data/Data_train/valbio_min_max_par_especes.csv')
```
