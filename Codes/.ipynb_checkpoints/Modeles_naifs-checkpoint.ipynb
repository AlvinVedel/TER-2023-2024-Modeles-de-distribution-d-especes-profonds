```python
import pandas as pd
```

```python
blind = pd.read_csv("../Data/Data_submission/test_blind.csv", sep=";") # Jeu de données pour la prédiction et soumission Kaggle
#sparound = pd.read_csv("../Data/test_spAround_1km.csv", sep=";")
#enviro = pd.read_csv("../Data/enviroTab_pa_test.csv", sep=";")
presences = pd.read_excel("../Data/Data_train/Presences_Absences_train.xlsx") # Jeu de données pour les infos présences absences
```

```python
# Affichage jeu de donnée
blind
```

```python
# Comptage du nombre d'occurence de chaque espèce dans le jeu de données des présences
# -> établir espèces les plus présentes 
nb_especes = presences["speciesId"] 
nb_especes = nb_especes.to_frame()
nb_especes['count'] = 1
sumup = nb_especes.groupby('speciesId').sum()
sumup = sumup.sort_values("count", ascending=False)
sumup
```

```python
top_100= []
# On extrait les 50 plus présentes et on récupère leurs Ids
i=0
while i< 100 :
    value = sumup.iloc[i]["count"]
    top_100 += sumup[sumup["count"]==value].index.tolist()
    i+=1
```

```python
# Construction de la variable str 'predict' qui sera associé aux ids du csv blind pour soumission Kaggle 
predict = ''
for i in top_100 :
    predict +=str(i)+ " "
```

```python
# Ajout colonne Predicted à une copie de blind
most_present = blind.copy()
most_present["Predicted"] = predict
```

```python
# Construction du csv de soumission avec uniquement colonnes utiles
most_present_test = most_present[["Id", "Predicted"]]
most_present_test
```

```python
# Modification format pour suppression des indices de lignes
most = most_present_test.groupby(by="Id").sum()
most
```

```python
# Ecriture du df en csv
most.to_csv("most_present_test2.csv")
```

```python
import random
# Définition de la fonction random_species
def random_species(df):
    """
    df = jeu de donnée train
    
    La fonction a pour but de retourner 5 espèce aléatoirement parmi les N présentes dans le dataframe
    -> Retour des 5 espèces en chaine de caractère
    """
    random_5= []
            # On extrait les 5 valeurs au hasard et on récupère leurs Ids
    i=0
    nb_species = len(df.index)
    while i< 5 :
        alea = random.randint(0, nb_species-1) 
        value = df.iloc[alea]
        if value.name not in random_5 :  
            random_5.append(value.name)
            i+=1

    predict = ''
    for i in random_5 :
        predict +=str(i)+ " "
    return predict
# fonction qui retourn 5 espèce aléatoires
```

```python
# test de la fonction 
random_species(sumup)
```

```python
# Création d'un df copie de blind avec une création de predicted aléatoire pour chaque ligne du df
random_test = blind.copy()
random_sp = []
for i in range (0, len(random_test.index)) :
    random_sp.append(random_species(sumup))

    
random_test["Predicted"] = random_sp
```

```python
# On isole les colonnes utiles de Id et predicted
random = random_test[['Id', 'Predicted']]
```

```python
# Transformation pour suppression des indices de lignes et écriture csv
randoms = random.groupby(by="Id").sum()
randoms.to_csv("random_test.csv")
```
